```python
# Mengecek data yang memiliki missing value untuk setiap kolom
data[data.isnull().any(axis=1)]
```

```python
# Mengubah object menjadi numeric menggunakan LabelEncoder
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
for i in data.select_dtypes(include='object').columns:
    data[i][data[i].notnull()] = encode.fit_transform(data[i])
data.head()
```

```python
#Normalisasi data dengan MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
sca = MinMaxScaler()
data_sca = pd.DataFrame(sca.fit_transform(data),columns=data.columns)
data_sca.head()
```

```python
# drop row yang punya missing value
new_data = data_sca.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# menampilkan jumlah missing value setelah di drop
new_data.isnull().sum()
```

```python
new_data.head()
```

```python
# Membuat fitur dan target
X = new_data.drop(['species'], axis=1)
y = new_data['species']
# species menjadi target karena kita ingin memprediksi species dari penguin
```

```python
# split data menjadi train dan test/valid
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)
print('Total X_train', len(X_train))
print('Total X_test', len(X_valid))
print('Ttal y_train', len(y_train))
print('Total y_test', len(y_valid))
```

```python
# exercise missing values:
# 1. load the data
# 2. check if there are missing values
# 3. print total missing values for each column
# 4. handle missing values by using several methods mention above
# 5. compare your result and anlyze it


# for comparation you can use bellow code
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)

# Evaluate the random forest regressor model using the score_dataset function
mae = score_dataset(X_train, X_valid, y_train, y_valid)

print('Mean Absolute Error:', mae)
```
